## The `.astype()` Method

Allows us to convert the data type of a series from one to another, this method requires a series to not have any null values


```python
nba = pd.read_csv(nbacsv).dropna(how = "all")
nba["Salary"].fillna(0, inplace = True)
nba["College"].fillna("None", inplace = True)
nba.head(6)
```

```python
nba.dtypes
nba.info()
```

```python
nba["Salary"] = nba["Salary"].astype("int")
```

```python
nba.head(3)
```

```python
nba["Number"] = nba["Number"].astype("int")
nba["Age"] = nba["Age"].astype("int")
nba.head(3)
```

## Rank Values with the `.rank()` Method

To use the Rank method,i have to get rid of all the null values. 

```python
nba = pd.read_csv(nbacsv).dropna(how = "all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head(3)
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
nba.head(3)
```

```python
nba.sort_values(by = "Salary", ascending = False)
```
