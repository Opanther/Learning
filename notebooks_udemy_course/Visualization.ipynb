```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline 
```

## Use the `plot` Method to Render a Line Chart

```python
bb = data.DataReader(name = "BB", data_source = "yahoo", start = "2007-01-01", end = "2020-12-31") # stock data for blackberry
bb.head(3)
```

```python
bb.plot(y = "High") # the plot method 
```

```python
bb["High"].plot() # method to extract the high and then plot the series
```

```python
bb[["High", "Close"]].plot() # extraction of a subset of series and plot
```

## Modifying Plot Aesthetics with `matplotlib` Templates

## Creating Bar Graphs to Show Counts

```python
def rank_performance(stock_price):
    if stock_price <= 10:
        return "Poor"
    elif stock_price <= 50:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
plt.style.use("ggplot")
# bb["Close"].apply(rank_performance).value_counts().plot(kind = "bar")
bb["Close"].apply(rank_performance).value_counts().plot(kind = "barh") # use kind to define type of graph
```

## Creating Pie Charts to Represent Proportions

```python
bb = data.DataReader(name = "BB", data_source = "yahoo", start = "2007-01-01", end = "2020-12-31")
bb.head(3)
```

```python
bb["Close"].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 30.55494908457586:
        return "Above Average"
    else:
        return "Below Average"
```

```python
plt.style.use("dark_background")
bb["Close"].apply(rank_performance).value_counts().plot(kind = "pie", legend = True)
```
