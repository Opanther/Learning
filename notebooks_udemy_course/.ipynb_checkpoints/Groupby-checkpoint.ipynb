## Methods on the Groupby Object and `DataFrame` Columns

```python
sectors.max() # returns the company that appears last within each group
sectors.min() # returns the company that appears first within each group
sectors.sum() # only applies to numeric columns
sectors.mean()
```

```python
sectors["Revenue"].sum() # use column name 
sectors["Employees"].sum()
sectors["Profits"].max()
sectors["Profits"].min()
sectors["Employees"].mean()

sectors[["Revenue", "Profits"]].sum()
```

## Grouping by Multiple Columns

```python
fortune = pd.read_csv(data, index_col = "Rank")
sectors = fortune.groupby(["Sector", "Industry"]) # Feed a list 
fortune.head(3)
```

```python
sectors.size()
sectors.sum()
sectors["Revenue"].sum()
sectors["Employees"].mean()
```

## The `.agg()` Method

```python
fortune = pd.read_csv(data, index_col = "Rank")
sectors = fortune.groupby("Sector")
fortune.head(3)
```

```python
sectors["Employees"].mean()
```

```python
sectors.agg({"Revenue" : ["sum", "mean"],
             "Profits" : "sum",
              "Employees" : "mean"}) #with agg method, i can specify the columns and what operations i want to perform on each, use dict: column = key and value = operation
```

```python
sectors.agg(["size", "sum", "mean"])
```

## Iterating through Groups

```python
fortune = pd.read_csv(data, index_col = "Rank")
sectors = fortune.groupby("Sector")
fortune.head(3)
```

```python
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for sector, data_f in sectors:
    highest_revenue_company_in_group = data_f.nlargest(1, "Revenue")
    df = df.append(highest_revenue_company_in_group)
```

```python
df
```

```python
cities = fortune.groupby("Location")
df = pd.DataFrame(columns = fortune.columns) # empty df to store results 
df
```

```python
for city, data in cities:
    highest_revenue_in_city = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_in_city)
```
