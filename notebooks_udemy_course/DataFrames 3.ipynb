## Rename Index Labels or Columns in a `DataFrame`

```python
bond = pd.read_csv(r"C:\Users\Gebruiker\Documents\Python Scripts\notebooks_udemy_course\pandas\jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
#Look at Doc for Rename method to understand
# Four methods to change rows name 
bond.rename(mapper= {"GoldenEye":"Golden Eye", 
                     "The World Is Not Enough": "Best Bond Movie Ever"}) # This operation is non-mutational

bond.rename(mapper= {"GoldenEye":"Golden Eye", 
                     "The World Is Not Enough": "Best Bond Movie Ever"}, axis = 0)
bond.rename(mapper= {"GoldenEye":"Golden Eye", 
                     "The World Is Not Enough": "Best Bond Movie Ever"}, axis = "rows")
bond.rename(mapper= {"GoldenEye":"Golden Eye", 
                     "The World Is Not Enough": "Best Bond Movie Ever"}, axis = "index")
#Instead of Using Mapper we can use the index argument
bond.rename(index= {"GoldenEye":"Golden Eye", 
                     "The World Is Not Enough": "Best Bond Movie Ever"}, inplace = True)
bond
```

```python
#Four Methods to change column name 
bond.rename(mapper = {"Year" : "Release Date", "Box Office" : "Revenue"}, axis=1)
bond.rename(mapper = {"Year" : "Release Date", "Box Office" : "Revenue"}, axis="columns")
bond.rename(columns = {"Year" : "Release Date", "Box Office" : "Revenue"})
bond.rename(columns = {"Year" : "Release Date", "Box Office" : "Revenue"}, inplace = True)
bond
```

```python
bond.head(1)
```

```python
#Another Strategy to rename columns, helpful if i want to rename all columns at once, one can overwrite the attribute with a list of new strings value, the length of the list has to be identical to the length of the original list

bond.columns = ["Year of Release", "Actor", "Director", "Gross", "Cost", "Salary"]
bond
```

## The `.query()` Method
Make sure to check that the Column names don't have any spaces, if they do, one have to run an operation to replace those spaces, with a connecting character. 

```python
bond = pd.read_csv(r"C:\Users\Gebruiker\Documents\Python Scripts\notebooks_udemy_course\pandas\jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
bond.columns = [column_name.replace(" ", "_") for column_name in bond.columns] # list comprehension to replace white space with underscores
bond.head(1)
```

```python
# Documentation say that the query method tends to be faster
bond.query('Actor == "Sean Connery"')
bond.query("Director == 'Terence Young'") # Using different quoting style
bond.query("Actor != 'Roger Moore'")
```

```python
bond.query("Box_Office > 600")
```

```python
bond.query("Actor == 'Roger Moore' or Director == 'John Glen'") # allows query to be more human and readable
```

```python
bond.query("Actor in ['Timothy Dalton', 'George Lazenby']")

bond.query("Actor not in ['Sean Connery', 'Roger Moore']")
```

## A Review of the `.apply()` Method on Single Columns

```python
bond = pd.read_csv(r"C:\Users\Gebruiker\Documents\Python Scripts\notebooks_udemy_course\pandas\jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + " MILLIONS!"
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions)
```

```python
bond.head(3)
```

## The `.apply()` Method with Row Values

```python
bond = pd.read_csv(r"C:\Users\Gebruiker\Documents\Python Scripts\notebooks_udemy_course\pandas\jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue"
    
bond.apply(good_movie, axis = "columns")
```
