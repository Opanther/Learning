```python
df = pd.read_csv(r"C:\Users\Gebruiker\Documents\Python Scripts\notebooks_udemy_course\pandas\employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

## Filter with More than One Condition (AND)

```python
#Define conditions separately.
mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "Marketing"

df[mask1 & mask2]
```

## Filter with More than One Condition (OR)

```python
mask1 = df["Senior Management"]
mask2 = df["Start Date"] < "1990-01-01"

df[mask1 | mask2] 
```

```python
mask1 = df["First Name"] == "Robert"
mask2 = df["Team"] == "Client Services"
mask3 = df["Start Date"] > "2016-06-01"

df[(mask1 & mask2) | mask3]
```

## The `.isin()` Method

Useful when comparing multiple conditions

```python
mask1 = df["Team"] == "Legal"
mask2 = df["Team"] == "Sales"
mask3 = df["Team"] == "Product"

df[mask1 | mask2 | mask3]
```

```python
mask = df["Team"].isin(["Legal", "Sales", "Product"])
df[mask]
```

## The `.isnull()` and `.notnull()` Methods

```python
mask = df["Team"].isnull()

df[mask]
```

```python
condition = df["Gender"].notnull()

df[condition]
```
